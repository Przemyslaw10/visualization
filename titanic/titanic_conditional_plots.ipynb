```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

titanic = pd.read_csv('train.csv', usecols=["Survived", "Pclass", "Sex",
                                             "Age", "SibSp", "Parch", "Fare", "Embarked"])
print(titanic.shape[0])
titanic.dropna(inplace=True)
print(titanic.shape[0])
```

```python
sns.distplot(titanic['Age'])
plt.show()
```

```python
g = sns.FacetGrid(titanic, col="Survived", row="Pclass", hue="Sex", size=5)
g.map(sns.kdeplot, "Age", shade=True, legend=True)
g.add_legend()
sns.despine(left=True, bottom=True)
plt.show()
```
